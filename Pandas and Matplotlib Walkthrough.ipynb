## Little crash course about Pandas 

```python
import pandas as pd
```

** import csv file **

```python
banks = pd.read_csv('banklist.csv')

# 存檔用 pd.write_csv()
```

** show the head of csv file **

```python
banks.head()
```

** Delete columns named 'Updated Date' **

```python
banks.drop(columns = ['Updated Date'])
```

```python
banks.head()
```

** show columns of csv file **

```python
banks.columns
```

** show all states(ST) in csv file **

```python
banks['ST'].unique()
```

** calculate unique states number  **

```python
banks['ST'].nunique()
```

** with groupby function, count how many banks fall in each states and show the first five rows **

```python
banks.groupby("ST").count().sort_values('Bank Name',ascending=False).iloc[:5]
```

** What are the top 5 states with the most failed banks?  **

```python
banks.groupby("ST").count().sort_values('Bank Name',ascending=False).iloc[:5]['Bank Name']
```

** What are the top 5 acquiring institutions?  **

```python
banks['Acquiring Institution'].value_counts().iloc[:5]
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

# subplot('行數', '列數', '第幾個')

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
```
